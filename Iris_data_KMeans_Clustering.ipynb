### Importing StandardScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
std=StandardScaler()
```

### Selecting the features to Standardize

### KMeans Clustering

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=3,random_state=1)
```

```python
kmeans.fit_transform(df_std)
```

```python
kmeans.labels_
```

```python
df.head()
```

```python
df['Clusters']=kmeans.labels_
```

```python
df['Clusters'].value_counts()
```

```python
df_plot=df.drop(['Id','Species'],axis=1)
```

```python
sns.pairplot(df_plot,hue='Clusters')
plt.show()
```

### Elbow Method

```python
wss=[]
cluster_list=list(range(1,6))
for x in cluster_list:
    kmeans=KMeans(n_clusters=x,init='k-means++',max_iter=300,n_init=10,random_state=1)
    kmeans.fit(df_std)
    wss.append(kmeans.inertia_)
```

```python
### Plot the elbow method

plt.plot(cluster_list,wss)
plt.title('Elbow_Method',fontsize=20)
plt.xlabel('No. of Clusters')
plt.ylabel('wss')
plt.show()
```
